# QUESTION 2
# Max, min , mean of asked amount,asked equity, asked valuation? 

```python
import pandas as pd

# Assuming you have already read the CSV file into a DataFrame
data = pd.read_csv('EDA_Shark_Tank_India.csv')

# Find the maximum, minimum, and mean values for 'Asked Amount'
max_asked_amount = data['pitcher_ask_amount'].max()
min_asked_amount = data['pitcher_ask_amount'].min()
mean_asked_amount = data['pitcher_ask_amount'].mean()

# Find the maximum, minimum, and mean values for 'Asked Equity'
max_asked_equity = data['ask_equity'].max()
min_asked_equity = data['ask_equity'].min()
mean_asked_equity = data['ask_equity'].mean()

# Find the maximum, minimum, and mean values for 'Asked Valuation'
max_asked_valuation = data['ask_valuation'].max()
min_asked_valuation = data['ask_valuation'].min()
mean_asked_valuation = data['ask_valuation'].mean()

# Display the results
print("For Asked Amount:")
print(f"Max: {max_asked_amount}, Min: {min_asked_amount}, Mean: {mean_asked_amount}")

print("\nFor Asked Equity:")
print(f"Max: {max_asked_equity}, Min: {min_asked_equity}, Mean: {mean_asked_equity}")

print("\nFor Asked Valuation:")
print(f"Max: {max_asked_valuation}, Min: {min_asked_valuation}, Mean: {mean_asked_valuation}")
```

# QUESTION 3
# Max & Min asked- equity,asked-valuation and asked amount episode-wise? 

```python
import pandas as pd

# Assuming you have already read the CSV file into a DataFrame
data = pd.read_csv('EDA_Shark_Tank_India.csv')

# Grouping data by 'Episode' and finding max and min values for each episode
max_values_episode_wise = data.groupby('episode_number').agg({'ask_equity': 'max', 'ask_valuation': 'max', 'pitcher_ask_amount': 'max'})
min_values_episode_wise = data.groupby('episode_number').agg({'ask_equity': 'min', 'ask_valuation': 'min', 'pitcher_ask_amount': 'min'})

# Displaying the results
print("Maximum Values Episode-wise:")
print(max_values_episode_wise)

print("\nMinimum Values Episode-wise:")
print(min_values_episode_wise)
```

# QUESTION 5
# Episode wise minimum and maximum of sharks invested? 

```python
import pandas as pd

# Assuming you have already read the CSV file into a DataFrame
data = pd.read_csv('EDA_Shark_Tank_India.csv')

# Grouping data by 'Episode' and finding min and max values for sharks invested
min_sharks_per_episode = data.groupby('episode_number')['total_sharks_invested'].min()
max_sharks_per_episode = data.groupby('episode_number')['total_sharks_invested'].max()

# Displaying the results
print("Episode-wise Minimum Sharks Invested:")
print(min_sharks_per_episode)

print("\nEpisode-wise Maximum Sharks Invested:")
print(max_sharks_per_episode)
```

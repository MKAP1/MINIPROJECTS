# QUESTION 1
# What is the maximum number of matches played by an individual player in a season?

```python
import pandas as pd


data = pd.read_csv('IPL_Dataset.csv')


max_matches_per_player = data.groupby('Player')['Mat'].max()


max_matches_player = max_matches_per_player.idxmax()
max_matches_count = max_matches_per_player.max()
print(max_matches_player)
print(max_matches_count)
print(f"The maximum number of matches played by an individual player in a season is {max_matches_count} by {max_matches_player}.")
```

# QUESTION 2
# What is the highest average of a player in a season?

```python
import pandas as pd


data = pd.read_csv('IPL_Dataset.csv')


max_avg_per_player = data.groupby('Player')['Avg'].max()


max_avg_player = max_avg_per_player.idxmax()
max_avg_value = max_avg_per_player.max()

print(f"The highest batting average of an individual player in a season is {max_avg_value} by {max_avg_player}.")
```

# QUESTION 3
# What is the average number of matches played by the players in a season?

```python
import pandas as pd


data = pd.read_csv('IPL_Dataset.csv')


average_matches_per_player = data.groupby('Player')['Mat'].mean()


overall_average_matches = average_matches_per_player.mean()

print(f"The average number of matches played by players in a season is {overall_average_matches:.2f}.")
```

# QUESTION 11
# Plot a histogram of number of matches played in a season by players.


```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('IPL_Dataset.csv')


plt.figure(figsize=(10, 6))
plt.hist(data['Mat'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Number of Matches Played by Players')
plt.xlabel('Number of Matches Played')
plt.ylabel('Number of Players')
plt.grid(axis='y', alpha=0.75)

plt.show()
```

# QUESTION 14
# Bar chart for top 20 players with best strike rate in a season.


```python
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv('IPL_Dataset.csv')


top_20_strike_rate_players = data.nlargest(20, 'SR')


plt.figure(figsize=(12, 8))
plt.bar(top_20_strike_rate_players['Player'], top_20_strike_rate_players['SR'], color='green')
plt.title('Top 20 Players with the Best SR in a Season')
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

# KDE plot for averages on a season.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv('IPL_Dataset.csv')


plt.figure(figsize=(10, 6))
sns.kdeplot(data['Avg'], fill=True, color='purple')
plt.title('Kernel Density Estimate (KDE) Plot for Averages in a Season')
plt.xlabel('Average')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
